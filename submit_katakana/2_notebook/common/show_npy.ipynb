```python
import numpy as np
import matplotlib.pyplot as plt

img_n = 100

img = np.load('../../1_data/train_data.npy', allow_pickle=True)
label = np.load('../../1_data/train_label.npy', allow_pickle=True)

plt.figure(figsize=(1, 1))
for i in range(img_n):
    plt.axis('off')
    plt.title(label[i ])
    plt.imshow(img[i][0], cmap='gray')
    plt.show()
```
